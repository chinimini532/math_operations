```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('poverty.csv', sep='\t')
y = df['Brth15to17'].values
x= df['PovPct'].values

# Parameters for Gradient Descent
m, c = 0, 0            # Initial values of m and c
L = 0.0001             # Learning rate
iterations = 1000      # Maximum number of iterations
n = len(y)             # Number of data points
threshold = 0.001     # Convergence threshold for the cost

# Gradient Descent Algorithm
for i in range(iterations):
    # Calculate predictions and errors
    y_pred = m * x + c
    error = y_pred - y
    
    # Compute gradients
    dm = (2 / n) * np.sum(error * x)
    dc = (2 / n) * np.sum(error)
    
    # Update parameters
    m = m - L * dm
    c = c - L * dc
    
    # Calculate cost (Mean Squared Error)
    cost = np.mean(error**2)
    
    # Print progress every 100 iterations
    if i % 100 == 0:
        print(f"Iteration {i+1}: m = {m}, c = {c}, cost = {cost}")
    
    # Stop if cost is less than the threshold
    if cost < threshold:
        print(f"Converged after {i+1} iterations with cost = {cost}")
        break

# Display final values of m and c
print(f"Optimal m: {m}, Optimal c: {c}")

# Plot the results
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x, m * x + c, color='red', label='Fitted line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression using Gradient Descent')
plt.show()
```
